```python
import pandas as pd

# top 100 games in google play store games in 2021
google_game = pd.read_csv("android-games.csv")

# top 100 games in steam in 2021
steam_game = pd.read_csv("steam_top_100.csv")
```

```python
google_game.head()
```

```python
steam_game.head()
```

```python
gg_set = set(google_game['title'])
sg_set = set(steam_game['Game'])
```

```python
print("交集：同時為google play與steam遊戲評價前一百名的遊戲：")

# title of games in top 100 google play games and top 100 steam games
gg_set & sg_set
```

```python
# title of games that are not 100 google game but top 100 steam games
set1 = sg_set - gg_set

print(f"差集：steam遊戲評價前一百名，但不在google play評價前一百名的遊戲：\n（共{len(set1)}個，只印出十個）")
# just display 10 items in the set because there are too many
for inx, item in enumerate(set1):
    if inx > 10: break
    print(item)
```

```python
# title of games that are top 100 google games or steam games
set2 = gg_set | sg_set

print(f"聯集：steam遊戲評價前一百名或在google play評價前一百名的遊戲：\n（共{len(set2)}個，只印出十個）")
# just display 10 items in the set because there are too many
for inx, item in enumerate(set2):
    if inx > 10: break
    print(item)
```
